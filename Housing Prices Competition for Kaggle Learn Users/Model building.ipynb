```python
import pandas as pd
# We don't import ALL of sklearn cuz that's too heavy
# import sklearn
```

```python
data = pd.read_csv("DATA/train.csv")
```

```python
data
```

```python
data.columns
```

```python
features = ["LotFrontage", "LotArea"]
```

```python
X = data[features]

# This other one does the same but is more flexible
# As to take less rows, etc.
# X = data.loc[:, features]
```

```python
X.describe()
```

### Here we find different counts for both features. <br>There's plenty of ways to go about this but I will, for now, just drop said rows

```python
X = X.loc[X["LotFrontage"].notna()]
X.describe()
```

```python
X.index
```

### This here work as an... index... for Y <br> So I'll take that same slice to stay in accordance with X

```python
y = data.SalePrice[X.index]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error

from sklearn.tree import DecisionTreeRegressor
```

```python
# We set a train-test split so we can check if each model holds
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=1)
```

### First a DTR for small testing

```python
# We define the model. 
# We set a number for random_state to ensure same results each run
housing_prices_DTR = DecisionTreeRegressor(random_state=1)

# Fit model
housing_prices_DTR.fit(X_train, y_train)
```

```python
# It's good practice to just set a dict to create a df

DTR_preds = housing_prices_DTR.predict(X_test)

d = {"Predictions": DTR_preds,
    "Target": y_test}
DTR_comparison = pd.DataFrame(data=d)

print("Mean absolute error:", mean_absolute_error(y_test, DTR_preds))
```

```python
# Side-by-side comparison
DTR_comparison
```

# Now this model SUCKS
### It's highly overfitted, so we'll set up a new one
## Random forests

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
# We define the model. 
# We set a number for random_state to ensure same results each run
housing_prices_RFR = RandomForestRegressor(random_state=1)

# Fit model
housing_prices_RFR.fit(X_train, y_train)
```

```python
# It's good practice to just set a dict to create a df

RFR_preds = housing_prices_RFR.predict(X_test)

d = {"Predictions": RFR_preds,
    "Target": y_test}
RFR_comparison = pd.DataFrame(data=d)

print("Mean absolute error:", mean_absolute_error(y_test, RFR_preds))
```

```python
RFR_comparison
```
